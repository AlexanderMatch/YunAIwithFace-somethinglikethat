```python
from google.cloud import texttospeech
import IPython.display as ipd

# 初始化客戶端
client = texttospeech.TextToSpeechClient()
# 配置語音合成參數
text = input("請輸入要朗讀的文字: ")
synthesis_input = texttospeech.SynthesisInput(text=text)

# 設定情緒
emotion = ["default","angry","sad","happy"]
emotion_Setting = int(input())

# 選擇語音
voice = texttospeech.VoiceSelectionParams(
    language_code="en-AU",
    name="en-AU-Wavenet-A",
    ssml_gender=texttospeech.SsmlVoiceGender.FEMALE
)

# 設定音頻配置
match emotion[emotion_Setting]:
  case "default":
    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.LINEAR16,
        speaking_rate=1,  # 語速 大於1會變快
        pitch=0,          # 音高 -20.0 到 20.0
        volume_gain_db=-1  # 音量增益 -96.0 到 16.0
    )

  case "angry":
    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.LINEAR16,
        speaking_rate=0.9,  # 語速
        pitch=1.7,          # 音高
        volume_gain_db=8  # 音量增益
    )

  case "sad":
    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.LINEAR16,
        speaking_rate=0.8,  # 語速
        pitch=-0.2,          # 音高
        volume_gain_db=-0.5  # 音量增益
    )

  case "happy":
    audio_config = texttospeech.AudioConfig(
        audio_encoding=texttospeech.AudioEncoding.LINEAR16,
        speaking_rate=1.12,  # 語速
        pitch=3,          # 音高
        volume_gain_db=5  # 音量增益
    )

# 調用 API 合成語音
response = client.synthesize_speech(input=synthesis_input, voice=voice, audio_config=audio_config)

ipd.Audio(response.audio_content, autoplay=True)

# 保存語音為 MP3
#with open("output.mp3", "wb") as out:
#    out.write(response.audio_content)

# 播放語音
#import IPython.display as ipd
#ipd.Audio("output.mp3")
```
